```python
import names
import os
import pandas as pd
import random
from datetime import datetime, date, timedelta
import datetime

# customer table

df_cities = pd.read_csv("Desktop/Personal Project/gb.csv")

df_branches = df_cities.head(20)

df_cust = pd.DataFrame(columns = ["uid","name","branch"])


for i in range(200):
    df_len = len(df_cust)
    df_temp = df_branches.sample()
    df_temp = df_temp.reset_index()
    temp = df_temp.loc[0,"city"]
    
    df_cust.loc[df_len,"uid"] = df_len
    df_cust.loc[df_len,"name"] = names.get_full_name()
    df_cust.loc[df_len,"branch"] = temp
```

```python
# books
from random_word import RandomWords
r = RandomWords()

types_fiction = ["SCI-FI", "FANTASY", "ADVENTURE", "MYSTERY", "CRIME", "HISTORY", "ROMANCE", "HORROR"]
types_non_fic = ["BIOGRAPHY", "SELF-HELP", "HISTORY", "TRAVEL", "COOKING", "CRIME", "REFERENCE"]
genra_list = ["FICTION","NON-FICTION"]

df_books = pd.DataFrame(columns=["title","author","genre","subgenre"])

for j in range(200):
    print(j)
    
    genra = random.choice(genra_list)
    if genra == "FICTION":
        sub = random.choice(types_fiction)
    else:
        sub = random.choice(types_non_fic)
        genra = "NON-FICTION"
    
    df_len = len(df_books)
    df_books.loc[df_len,"genre"] = genra
    df_books.loc[df_len,"subgenre"] = sub
    df_books.loc[df_len,"title"] = str(r.get_random_word()+" "+r.get_random_word())
    df_books.loc[df_len,"author"] = names.get_full_name()
    
```

```python
# copies

df_copies = pd.DataFrame(columns=["unquie","title","author","branch"])

for k in range(400):
    df_temp = df_cust.sample()
    df_temp = df_temp.reset_index()
    temp = df_temp.loc[0,"branch"]
    
    df_temp_two = df_books.sample()
    df_temp_two = df_temp_two.reset_index()
    print(df_temp_two)
    
    if (k%5 == 0):
        r = 0
        for r in range(random.randint(1,14)):
            df_len = len(df_copies)
            df_copies.loc[df_len,"unquie"] = df_len+1337
            df_copies.loc[df_len,"title"] = df_temp_two.loc[0,"title"] 
            df_copies.loc[df_len,"author"] = df_temp_two.loc[0,"author"] 
            df_copies.loc[df_len,"branch"] = temp


    df_len = len(df_copies)
    df_copies.loc[df_len,"unquie"] = df_len+1337
    df_copies.loc[df_len,"title"] = df_temp_two.loc[0,"title"] 
    df_copies.loc[df_len,"author"] = df_temp_two.loc[0,"author"] 
    df_copies.loc[df_len,"branch"] = temp
    
#     print(df_copies.loc[df_len])
    
```

```python
# loans
startframe = pd.Timestamp(year = 2022,  month = 11, day = 20,tz = 'GMT') 

endframe = pd.Timestamp(year = 2023,  month = 2, day = 20,tz = 'GMT') 

df_loans = pd.DataFrame(columns = ["uid","ref_to_copy_id","custid","startdate","enddate"])

for p in range(len(df_copies)):
    startlocal = startframe
    
    while(startlocal < endframe):
        endlocal = startlocal + timedelta(days=random.randint(2,31))
        
        df_temp = df_cust.sample()
        df_temp = df_temp.reset_index()
        custid = df_temp.loc[0,"uid"]

        df_len = len(df_loans)
        df_loans.loc[df_len,"uid"] = str("loan_"+str(df_len*2))
        df_loans.loc[df_len,"ref_to_copy_id"] = df_copies.loc[p,"unquie"]
        df_loans.loc[df_len,"custid"] = custid
        df_loans.loc[df_len,"startdate"] = startlocal.date()
        df_loans.loc[df_len,"enddate"] = endlocal.date()
        
        startlocal = endlocal+ timedelta(days=random.randint(12,50))
```
